```python
import numpy as np
import pandas as pd
```

pivot tables

```python
data = {
    "Region":      ["North", "South", "North", "East", "South", "East", "North", "East", "South", "North"],
    "Salesperson": ["Alice", "Bob",   "Alice", "Carol", "Bob",   "Carol", "Dave",  "Dave",  "Alice", "Bob"],
    "Product":     ["A",     "B",     "B",     "A",     "A",     "B",     "A",     "B",     "A",     "B"],
    "Date":        ["2024-01-15", "2024-01-20", "2024-04-10", "2024-01-18",
                    "2024-05-22", "2024-02-14", "2024-06-30", "2024-03-05",
                    "2024-05-18", "2024-02-28"],          # ✅ Date column added
    "Sales":       [200, 150, 300, 180, 220, 130, 260, 190, 210, 170],
    "Units":       [10,   8,  15,   9,  11,   7,  13,  10,  11,   9],
}

df = pd.DataFrame(data)

# ✅ Convert 'Date' string to datetime first
df['Date'] = pd.to_datetime(df['Date'])

# Now these will work fine
df['Month']   = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)

print(df)
```

```python
pd.pivot_table(df,values = 'Sales', index = 'Region', columns ="Product")
```

```python
pivot2 = pd.pivot_table(df, values=['Sales','Units'], index = 'Region', columns='Product')
pivot2
```
