```python
import numpy as np
import pandas as pd
```

### GroupBy

```python
data = {
    "Region":      ["North", "South", "North", "East", "South", "East", "North", "East"],
    "Salesperson": ["Alice", "Bob",   "Alice", "Carol", "Bob",   "Carol", "Dave",  "Dave"],
    "Product":     ["A",     "B",     "B",     "A",     "A",     "B",     "A",     "B"],
    "Sales":       [200,     150,     300,     180,     220,     130,     260,     190],
    "Units":       [10,      8,       15,      9,       11,      7,       13,      10],
}

df = pd.DataFrame(data)
```

```python
df
```

### GroupBy Product and calculate the sum of sales

```python
prdt = df.groupby('Product')
prdt
for i, v in prdt:
    print(i)
    print(v)
```

```python
prdt = df.groupby('Region')
prdt
for i, v in prdt:
    print(i)
    print(v)
```

### Group By salesPerson and calculate the sum of the sales

```python
prdt = df.groupby('Salesperson')['Sales'].sum()
prdt
```

```python
#Group by multiple columns
#Group by region and product
prdt = df.groupby(['Region','Product'])['Sales'].sum()
prdt
```
