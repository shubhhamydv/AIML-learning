```python
import numpy as np
import pandas as pd
```

### Merging 2 dataframes

```python
employees = pd.DataFrame({
    "EmployeeID": [1, 2, 3],
    "EmployeeName": ["Alice", "Bob", "Charlie"],
    "Department": ["IT", "HR", "Finance"]
})
# DataFrame 2: Salary information
salaries = pd.DataFrame({
    "EmployeeID": [1, 2, 3],
    "Salary": [50000, 60000, 70000],
    "Bonus": [5000, 7000, 8000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries)
```

### Concatination of 2 DataFrames

```python
df1 = pd.DataFrame({
    "EmployeeID": [1, 2],
    "Name": ["Alice", "Bob"],
    "Department": ["IT", "HR"]
})

df2 = pd.DataFrame({
    "EmployeeID": [3, 4],
    "Name": ["Charlie", "David"],
    "Department": ["Finance", "Marketing"]
})

print(df1)
print(df2)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis = 1)
```

### Joining 2 Data-Frames

```python
df1 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
     
}, index=[1, 2, 3])

df2 = pd.DataFrame({
     
    "Score": [88, 92, 80]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```
